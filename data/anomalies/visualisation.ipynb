```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_BASE_DIR = '/home/drx/PycharmProjects/MA/data/anomalies'

anomalies = ['amplitude-larger', 'amplitude-smaller', 'global-extremum', 'local-extremum', 'pattern-anomaly']
```

```python
fig, axs = plt.subplots(5,2, sharex='col', sharey='row', figsize=(16,12))
for i, anomaly in enumerate(anomalies):
    path = f'{DATA_BASE_DIR}/{anomaly}'
    train_df = pd.read_csv(f'{path}/train_no_anomaly.csv', usecols=['timestamp','value-0'])
    test_df = pd.read_csv(f'{path}/test.csv', usecols=['timestamp','value-0'])
    axs[i,0].plot(train_df['timestamp'], train_df['value-0'])
    axs[i,0].set_title(f'Train data for {anomaly}')
    axs[i,1].plot(test_df['timestamp'], test_df['value-0'])
    axs[i,1].set_title(f'Test data for {anomaly}')
    axs[i,0].set_ylim([-2.5,2.5])
    axs[i,0].set_ylabel("value")

axs[-1,0].set_xlabel('timestamp')
axs[-1,1].set_xlabel('timestamp')
plt.tight_layout()
# plt.savefig("sine_data.png")
plt.show()
```
