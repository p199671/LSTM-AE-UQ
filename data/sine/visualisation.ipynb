```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_BASE_DIR = '/home/drx/PycharmProjects/MA/data/sine'

noise_levels = ["00", "01", "10", "30", "50"]
hidden_dims = [50, 100]      # [50, 100]
encoding_dims = [40, 80]    # [40, 80]
window_sizes = [10, 25]     # [10, 25]
```

```python
fig, axs = plt.subplots(5,2, sharex='col', sharey='row', figsize=(16,12))
for i, noise_level in enumerate(noise_levels):
    path = f'{DATA_BASE_DIR}/sine-noise-{noise_level}%'
    train_df = pd.read_csv(f'{path}/train_no_anomaly.csv', usecols=['timestamp','value-0'])
    test_df = pd.read_csv(f'{path}/test_no_anomaly.csv', usecols=['timestamp','value-0'])
    axs[i,0].plot(train_df['timestamp'], train_df['value-0'])
    axs[i,0].set_title(f'Train data ({noise_level}% noise)')
    axs[i,1].plot(test_df['timestamp'], test_df['value-0'])
    axs[i,1].set_title(f'Test data ({noise_level}% noise)')
    axs[i,0].set_ylim([-2.5,2.5])
    axs[i,0].set_ylabel("value")

axs[-1,0].set_xlabel('timestamp')
axs[-1,1].set_xlabel('timestamp')
# plt.savefig("sine_data.png")
plt.show()
```
